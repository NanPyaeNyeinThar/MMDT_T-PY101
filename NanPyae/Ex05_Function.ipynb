Ex 3: Write a function called number_of_factors that takes an integer and returns how many
factors the number has.

```python
def number_of_factors(n):
  i = 1
  count = 0
  var = []

  while(i<n):
    if(n%i == 0):
      count+=1
      var.append(i)
    i+=1

  print(f"The factors of {n}: {count}")
  print(f"The factors are: {var}")


n = int(input("Enter the integer:"))
number_of_factors(n)
```

Ex4: Write a function called first_diff that is given two strings and returns the first location in
which the strings differ. If the strings are identical, it should return -1.Ignore the case in your function. It means h and H are the same.

```python
def first_diff(str1, str2):
  str1 = str1.lower()
  str2 = str2.lower()

  if str1 == str2:
    return -1

  str_length = min(len(str1), len(str2))
  for i in range(str_length):
    if str1[i] != str2[i]:
      return i

  return str_length

input1 = input("Enter frist string:")
input2 = input("Enter second string:")

result = first_diff(input1, input2)

if result == -1:
  print(f"The two strings are identical")
else:
  print(f"The two strings differed at the location {result}")
```

```python
def solve_equ(a,b,c,d,e,f):
  D = a*d - c*b
  if D == 0:
    print(f"The unique value does not exit.")
  else:
    x = round((e*d - b*f)/D,2)
    y = round((e - a*x)/b,2)
    print(f"The value of x: {x}")
    print(f"The value of y: {y}")

a = int(input("Enter the value of a:"))
b = int(input("Enter the value of b:"))
c = int(input("Enter the value of c:"))
d = int(input("Enter the value of d:"))
e = int(input("Enter the value of e:"))
f = int(input("Enter the value of f:"))
print("------------------")

solve_equ(a,b,c,d,e,f)

```

Ex7:  Write a function called change_case that given a string, returns a string with each upper  case letter replaced by a lower case letter and vice-versa

```python
def change_case(text):
  result = ""

  for i in range(len(text)):
    if text[i].isupper():
      result +=text[i].lower()
    else:
      result +=text[i].upper()

  print(f"The result is: {result}")

input_str = input("Enter the string:")
change_case(input_str)
```
